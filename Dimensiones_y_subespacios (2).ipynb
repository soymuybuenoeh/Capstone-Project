```python
# Calculate percentages for 'A' and 'D'
counts['Porcentaje A'] = (counts['Cantidad A'] / (counts['Cantidad A'] + counts['Cantidad D'])) * 100
counts['Porcentaje D'] = (counts['Cantidad D'] / (counts['Cantidad A'] + counts['Cantidad D'])) * 100

# Fill NaN values with 0 (if a category is not present in a comparison)
counts.fillna(0, inplace=True)

# Format percentage columns
pd.options.display.float_format = '{:.2f}%'.format

# Display the updated DataFrame
display(counts)
```

```python
def extraer_texto(row):
    # Se usa la columna "Relacion" renombrada previamente
    relacion = row["Relacion"]
    por_a = row["Porcentaje A"]
    por_d = row["Porcentaje D"]

    if por_a > por_d:
        # Si Porcentaje A es mayor, se extrae el texto entre el punto y el guion bajo
        partes = relacion.split('.')
        if len(partes) > 1:
            return partes[1].split('_')[0]
    elif por_d > por_a:
        # Si Porcentaje D es mayor, se extrae el texto antes del punto
        return relacion.split('.')[0]
    return ""  # En caso de empate o formato inesperado

# Aplicar la función a cada fila y almacenar el resultado en la nueva columna "Pais con mas saliencia"
counts["Pais"] = counts.apply(extraer_texto, axis=1)

# Mostrar el DataFrame final
display(counts)
```

```python
pair_results = []
for col in columns_of_interest.columns:
    # Suponemos que el nombre de la columna tiene el formato "pais1.pais2_s"
    base = col.split('_')[0]  # e.g., "ghana.luxembourg"
    try:
        c1, c2 = base.split('.')
    except Exception as e:
        print(f"Error al procesar el formato de la columna '{col}': {e}")
        continue

    c1_mapped = mapping.get(c1, c1.title())
    c2_mapped = mapping.get(c2, c2.title())

    # Buscar en df_merged usando comparaciones en minúsculas para mayor robustez
    row_c1 = df_merged.loc[df_merged["pais"].str.lower() == c1_mapped.lower()]
    row_c2 = df_merged.loc[df_merged["pais"].str.lower() == c2_mapped.lower()]

    # Si alguna de las filas no se encuentra, se omite este par
    if row_c1.empty or row_c2.empty:
        print(f"Mapping no encontrado para '{c1_mapped}' o '{c2_mapped}' en df_merged. Se omite el par '{col}'.")
        continue

    s1 = row_c1["salience"].values[0]
    s2 = row_c2["salience"].values[0]
    d1 = row_c1["dimensiones"].values[0]
    d2 = row_c2["dimensiones"].values[0]

    # Comprobar consistencia: si s1 > s2 esperamos d1 > d2, y viceversa
    if (s1 > s2 and d1 > d2) or (s2 > s1 and d2 > d1):
        consistency = True
    else:
        consistency = False

    pair_results.append({
        "Pair": f"{c1_mapped} vs {c2_mapped}",
        "Salience A": round(s1,2),
        "Salience D": round(s2,2),
        "Dimensions A": d1,
        "Dimensions D": d2,
        "Consistent": consistency
    })

df_pairs = pd.DataFrame(pair_results)
df_pairs
```
